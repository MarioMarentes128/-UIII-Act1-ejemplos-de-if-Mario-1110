#### Ejemplo de if

```python
# 1. Ejemplo de if simple (con operador relacional)
# El if simple ejecuta un bloque de código solo si la condición evaluada es verdadera
edad = 18 # Asigna el valor 18 a la variable edad [1, 15].
if edad >= 18: # Compara si edad es mayor o igual a 18 (operador relacional) [1, 16].
    print("Eres mayor de edad.") # Imprime el mensaje si la condición es True [1, 16].
```

```python
# 1. Ejemplo de if simple (con operador relacional)
# El if simple ejecuta un bloque de código solo si la condición evaluada es verdadera
edad = 18 # Asigna el valor 18 a la variable edad [1, 15].
if edad >= 18: # Compara si edad es mayor o igual a 18 (operador relacional) [1, 16].
    print("No mayor de edad.") # Imprime el mensaje si la condición es True [1, 16].
```

```python
# 2. Ejemplo complejo (combinando todos los operadores)
# Este ejemplo verifica si un año es bisiesto, lo cual requiere cálculos aritméticos y lógica compleja.
anio = 2024 # Asigna un año a la variable [1, 20].
# Condición: divisible por 4 y no por 100, O divisible por 400 [1, 16].
if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print(f"{anio} es un año bisiesto.") # Uso de f-string para formatear la salida [1, 24].
else:
    print(f"{anio} no es un año bisiesto.") # Resultado si la lógica booleana es False [1, 16].
```

```python
# 3. Ejemplo de if else (con operadores aritméticos y relacionales)
# Esta estructura permite elegir entre dos alternativas: una si la condición se cumple y otra si es falsa.
numero = 7 # Se define un número entero [1, 20].
if numero % 2 == 0: # Usa el operador aritmético módulo (%) y el relacional (==) [1, 5, 16].
    print("El número es par.") # Se ejecuta si el resto de dividir por 2 es cero [3, 12].
else: # Bloque opcional para manejar el caso contrario [1, 2].
    print("El número es impar.") # Se ejecuta si la condición del if fue False [3, 19].
```

```python
# 4. Ejemplo de if elif else (con operadores lógicos y relacionales)
# Se utiliza cuando existen múltiples condiciones que evaluar de forma sucesiva.
calificacion = 100 # Variable para almacenar una nota [1, 20].
# Evalúa si la nota está en un rango usando operadores relacionales y lógicos (and) [1, 3, 13, 22].
if calificacion >= 90 and calificacion <= 100:
    print("Excelente") # Se ejecuta si ambas condiciones son True [3, 13].
elif calificacion >= 80: # Segunda condición si la primera falló [1, 2].
    print("Muy bien") # Se ejecuta si la nota es 80 o más (pero menor a 90) [3, 21].
else: # Caso por defecto si ninguna de las anteriores se cumplió [1, 2].
    print("Necesitas mejorar") # Acción final para notas menores a 80 [3, 19].
```
